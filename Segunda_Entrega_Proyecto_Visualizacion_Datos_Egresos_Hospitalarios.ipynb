```python
Data['Fecha de ingreso'] = pd.to_datetime(Data['Fecha de ingreso'],format='%Y-%m-%d')
```

```python
Data['Año'] = Data['Fecha de ingreso'].dt.year.astype('string')
Data['Año'].value_counts()
```

```python
Data['Mes'] = Data['Fecha de ingreso'].dt.month
Data['Mes'].value_counts()
```

```python
import calendar
Data['NombreMes'] = Data['Mes'].apply(lambda x: calendar.month_name[x])
Data['NombreMes'].value_counts()
```

Gráfico de lineas

```python
Data_Colombiano = Data[(Data['Nacionalidad'] == "Colombiano")]
Data_Colombiano
```

```python
Conteo_Colombiano = pd.pivot_table(Data_Colombiano, 
                                values='Valor', 
                                index=['Fecha de Egreso'],
                                columns=['Nacionalidad'], 
                                aggfunc=np.sum)
Conteo_Colombiano = pd.DataFrame(Conteo_Colombiano)
Conteo_Colombiano.reset_index(inplace=True)
Conteo_Colombiano = Conteo_Colombiano.rename(columns = {'index':'Fechas','Colombiano':'Precio'})
Conteo_Colombiano.dtypes
```
